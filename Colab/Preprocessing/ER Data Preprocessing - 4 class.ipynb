```
import os
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
data_dir_list = os.chdir('/content/drive/My Drive/Breast Cancer Treatment/_ER/')
```

```
PATH = '/content/drive/My Drive/Breast Cancer Treatment/_ER/'
SAVE_PATH = '/content/drive/My Drive/Breast Cancer Treatment/Numpy Arrays/4 Class ER'
MASK_SAVE_PATH = '/content/drive/My Drive/Breast Cancer Treatment/Numpy Arrays/4 Class ER/Masks/'

img_path = PATH + 'Image'
strong_path = PATH + 'Strong'
intermediate_path = PATH + 'Intermediate'
weak_path = PATH + 'Weak'
```

```
def data_preparation(patientNo):
  imPath = img_path + '/' + patientNo
  sPath = strong_path + '/' + patientNo
  iPath = intermediate_path + '/' + patientNo
  wPath = weak_path + '/' + patientNo

  img_list = sorted(os.listdir(imPath))
  strong_list = sorted(os.listdir(sPath))
  intermediate_list = sorted(os.listdir(iPath))
  weak_list = sorted(os.listdir(wPath))

  Img = []
  Mask = []
  Full_Mask = []

  for i in range(len(img_list)):
    img_name = img_list[i].split('.')[0]
    img = cv2.imread(imPath + '/' + img_list[i])
    sMask = cv2.imread(sPath + '/' + strong_list[i])              # img_name + '.png'
    iMask = cv2.imread(iPath + '/' + intermediate_list[i])        # img_name + '.png'
    wMask = cv2.imread(wPath + '/' + weak_list[i])                # img_name + '.png'

    sMask  =cv2.cvtColor(sMask, cv2.COLOR_BGR2GRAY)
    ret, sMask = cv2.threshold(sMask, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)

    iMask  =cv2.cvtColor(iMask, cv2.COLOR_BGR2GRAY)
    ret, iMask = cv2.threshold(iMask, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)

    wMask  =cv2.cvtColor(wMask, cv2.COLOR_BGR2GRAY)
    ret, wMask = cv2.threshold(wMask, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)

    # Creating combined mask from individual masks
    # background = 0    weak = 85   intermediate = 170  strong = 255
    iMask = (iMask/255) * 170
    wMask = (wMask/255) * 85
    mask = sMask + iMask + wMask
    for i in range(mask.shape[0]):
      for j in range(mask.shape[1]):
        if mask[i][j] > 255:
          mask[i][j] = 0
    Full_Mask.append(mask)

    for r in range(6):
      row = r*240
      for c in range(8):
        col = c*240
        sliced_imageX = img[row:row+240,col:col+240,:]
        sliced_imageY = mask[row:row+240,col:col+240]
        Img.append(sliced_imageX)
        Mask.append(sliced_imageY)

  Img_data = np.array(Img)
  Mask_data = np.array(mask)
  print(Img_data.shape)
  Full_Mask =np.array(Full_Mask)
  print(Full_Mask.shape)
  np.save(MASK_SAVE_PATH + patientNo, Full_Mask)
  print(Mask_data.shape)
  np.save(SAVE_PATH + '/' + 'ER ' + patientNo + ' Images', Img_data)
  np.save(SAVE_PATH + '/' + 'ER ' + patientNo + ' Masks', Mask_data)
  print(patientNo + ' done.')     
```

```
for patientNo in os.listdir(img_path):
  data_preparation(patientNo)
  
```
