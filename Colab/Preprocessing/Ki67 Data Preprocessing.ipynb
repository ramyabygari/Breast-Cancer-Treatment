```
import os
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
data_dir_list = os.chdir('/content/drive/My Drive/Breast Cancer Treatment/_Ki67/')
```

```
PATH = '/content/drive/My Drive/Breast Cancer Treatment/_Ki67/'
SAVE_PATH = '/content/drive/My Drive/Breast Cancer Treatment/Numpy Arrays/Ki67/Masks'

img_path = PATH + '/Image'
ip_path = PATH + '/Immunopositive'
in_path = PATH + '/Immunonegative'
```

```
def data_preparation(patientNo):
  imPath = img_path + '/' + patientNo
  ipPath = ip_path + '/' + patientNo
  inPath = in_path + '/' + patientNo

  img_list = sorted(os.listdir(imPath))
  ip_list = sorted(os.listdir(ipPath))
  in_list = sorted(os.listdir(inPath))

  Img = []
  Mask = []

  for i in range(len(img_list)):
    img_name = img_list[i].split('.')[0]
    img = cv2.imread(imPath + '/' + img_list[i])
    ipMask = cv2.imread(ipPath + '/' + ip_list[i])              # img_name + '.png'
    inMask = cv2.imread(inPath + '/' + in_list[i])        # img_name + '.png'

    ipMask  =cv2.cvtColor(ipMask, cv2.COLOR_BGR2GRAY)
    ret, ipMask = cv2.threshold(ipMask, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)

    inMask  =cv2.cvtColor(inMask, cv2.COLOR_BGR2GRAY)
    ret, inMask = cv2.threshold(inMask, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)

    # Creating combined mask from individual masks
    ipMask = (ipMask/255) * 255
    inMask = (inMask/255) * 128
    mask   = ipMask + inMask
    for i in range(mask.shape[0]):
      for j in range(mask.shape[1]):
        if mask[i][j] > 255:
          mask[i][j] = 0

    # for r in range(6):
    #   row = r*240
    #   for c in range(8):
    #     col = c*240
    #     sliced_imageX = img[row:row+240,col:col+240,:]
    #     sliced_imageY = mask[row:row+240,col:col+240]
    Img.append(img)
    Mask.append(mask)

  Img_data = np.array(Img)
  Mask_data = np.array(Mask)
  print(Img_data.shape)
  print(Mask_data.shape)
  # np.save(SAVE_PATH + '/' + 'Ki67 ' + patientNo + ' Images', Img_data)
  np.save(SAVE_PATH + '/' + patientNo, Mask_data)
  print(patientNo + ' done.')     
```

```
for patientNo in os.listdir(img_path):
  data_preparation(patientNo)
```
