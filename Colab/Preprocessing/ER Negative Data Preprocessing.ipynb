```
import os
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
```

```
PATH = '/content/drive/My Drive/Breast Cancer Treatment/_ER/'
SAVE_PATH = '/content/drive/My Drive/Breast Cancer Treatment/Numpy Arrays/Negative Masks/'

img_path = PATH + '/Image'
negative_path = PATH + '/Negative'
```

```
def data_preparation(patientNo):
  imPath = img_path + '/' + patientNo
  nPath = negative_path + '/' + patientNo

  img_list = sorted(os.listdir(imPath))
  negative_list = sorted(os.listdir(nPath))

  Img = []
  Mask = []

  for i in range(len(img_list)):
    img_name = img_list[i].split('.')[0]
    img = cv2.imread(imPath + '/' + img_list[i])
    nMask = cv2.imread(nPath + '/' + negative_list[i])              # img_name + '.png'

    nMask  =cv2.cvtColor(nMask, cv2.COLOR_BGR2GRAY)
    ret, nMask = cv2.threshold(nMask, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)

    for r in range(6):
      row = r*240
      for c in range(8):
        col = c*240
        sliced_imageX = img[row:row+240,col:col+240,:]
        sliced_imageY = nMask[row:row+240,col:col+240]
        Img.append(sliced_imageX)
        Mask.append(sliced_imageY)

  Img_data = np.array(Img)
  Mask_data = np.array(Mask)
  print(Img_data.shape)
  print(Mask_data.shape)

  np.save(SAVE_PATH + '/' + 'ER ' + patientNo , Mask_data)
  print(patientNo + ' done.')     
```

```
patients = ['IHC 221','IHC 229', 'IHC 232', 'IHC 239', 'IHC 242','IHC 252', 'IHC 263']

for patientNo in patients:
  data_preparation(patientNo)
```
