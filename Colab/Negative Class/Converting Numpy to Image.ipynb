```
def writeImagesUtil(DEST_DIR,folderName,data):
  os.chdir(DEST_DIR)
  os.mkdir(folderName)
  os.chdir(DEST_DIR + folderName + '/')
  for j in range(len(data)):
    cv2.imwrite(DEST_DIR + folderName + '/' + str(j+1) + '.png', data[j])
```

```
def writeImages(SOURCE_PATH, EXTENDED_PATH):
  patients = sorted(os.listdir(SOURCE_PATH + EXTENDED_PATH))
  for i in range(len(patients)):
    filePath = SOURCE_PATH + EXTENDED_PATH + patients[i]
    data = np.load(filePath)
    folderName = patients[i].split('.')[0]
    writeImagesUtil(SAVE_PATH + EXTENDED_PATH, folderName,data)
```

```
writeImages(NP_PATH,'Train/')
writeImages(NP_PATH,'Test/')
```
