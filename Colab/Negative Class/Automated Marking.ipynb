```
import cv2
from google.colab.patches import cv2_imshow
from skimage import data
from skimage.color import rgb2hed, hed2rgb
import numpy as np
```

```
FOLDER = '/content/drive/My Drive/Breast Cancer Treatment/_ER/Image/'
SAVE_FOLDER = '/content/drive/My Drive/Breast Cancer Treatment/_ER/Negative/'
patient_no = 'IHC 242/'
image_no = '_10588'
extension = '.tif'
img = cv2.imread(FOLDER + patient_no + image_no + extension)
```

```
er = rgb2hed(img)
er = er[:,:,1]
normalized = cv2.normalize(er,None,alpha = 0, beta = 255, norm_type = cv2.NORM_MINMAX)
normalized[normalized <= 100] = 0
normalized[normalized >  100] = 255
```

```
import matplotlib.pyplot as plt
f = plt.figure(figsize=(20, 40))
plt.subplot(131).imshow(img)
plt.subplot(132).imshow(normalized,'gray')
```

```
cv2.imwrite(SAVE_FOLDER + patient_no + image_no + '.png',normalized)
```
